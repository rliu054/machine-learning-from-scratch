```python
from Regression import lin_reg

x_arr, y_arr = lin_reg.load_data_set('Data/ex0.txt')
print(x_arr[0:2], y_arr[0:2])
```

```python
ws = lin_reg.standard_reg(x_arr, y_arr)
ws
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x_mat = np.mat(x_arr)
y_mat = np.mat(y_arr)

y_hat = x_mat * ws
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x_mat[:, 1].flatten().A[0],
           y_mat.T[:, 0].flatten().A[0])

x_copy = x_mat.copy()
x_copy.sort(0)
y_hat = x_copy * ws
ax.plot(x_copy[:, 1], y_hat)
plt.show()
```
